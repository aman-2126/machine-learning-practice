```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import StandardScaler, Normalizer
```

```python
data_file = "./linear_regression/train.csv"
rawdata = pd.read_csv(data_file, names = ['a', 'b'])
rawdata = rawdata.dropna()
# print(rawdata)
```

```python
# dataset = datasets.load_diabetes()
# print(type(dataset))
```

```python
# print(dataset.keys())
rawdata_array = rawdata.to_numpy()
print(rawdata_array)
```

```python
# data_x = dataset.data[:, np.newaxis, 3]
# data_y = dataset.target
data_x = rawdata_array[:, 0:1]
data_y = rawdata_array[:, 1:2]
# print(data_x)
# print(data_y)
```

```python
# raw data
pd.DataFrame(data_x).plot(kind = 'density')
plt.show()
```

```python
# standardized data
scaler = StandardScaler().fit(data_x)
standardized_x = scaler.transform(data_x)
pd.DataFrame(standardized_x).plot(kind = 'density')
plt.show()

```

```python
# normalized data
scaler = Normalizer().fit(data_x)
normalized_x = scaler.transform(data_x)
pd.DataFrame(normalized_x).plot(kind = 'density')
plt.show()
```
